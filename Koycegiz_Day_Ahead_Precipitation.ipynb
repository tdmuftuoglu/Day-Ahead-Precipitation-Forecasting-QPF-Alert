```python
# ==============================================================================
# PRECIPITATION FORECASTING PIPELINE
# ==============================================================================
# Author: Dr. Tevfik Denizhan MUFTUOGLU
#
# License: MIT License
# Permission is hereby granted, free of charge, to any person obtaining a copy
# of this software and associated documentation files (the "Software"), to deal
# in the Software without restriction, including without limitation the rights
# to use, copy, modify, merge, publish, distribute, sublicense, and/or sell
# copies of the Software, and to permit persons to whom the Software is
# furnished to do so, subject to the following conditions:
#
# The above copyright notice and this permission notice shall be included in all
# copies or substantial portions of the Software.
# ==============================================================================
# DESCRIPTION:
# This script implements a dual-stage machine learning framework for daily
# precipitation forecasting. It performs the following operations:
# 1. Data Loading & Preprocessing: Splits data into chronological Train/Test sets.
# 2. Statistical Analysis: Generates descriptive statistics for data distribution validation.
# 3. Regression Modeling: Optimizes 7 ML models (CatBoost, LightGBM, XGBoost,
#    Random Forest, Gradient Boosting, Linear Regression, Huber Regressor) and
#    evaluates them using R2, RMSE, MAE, and MSE metrics.
# 4. Classification Modeling: Detects extreme precipitation events (>95th percentile).
# 5. Advanced Analysis: Conducts SHAP feature importance and Extreme Value Analysis (GEV).
# 6. Visualization: Generates publication-ready plots and time-series analysis.
# 7. Export: Saves all results to 'Analysis_Results.xlsx'.
# ==============================================================================

print("[STEP 0] Installing and Importing Libraries...")
!pip install -q prophet catboost pymannkendall shap calmap openpyxl

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from google.colab import files
import io
from IPython.display import display
import warnings
import time
warnings.filterwarnings("ignore")

# --- Libraries ---
from prophet import Prophet
import shap
from scipy.stats import genextreme as gev
import calmap
from sklearn.model_selection import train_test_split, RandomizedSearchCV, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler

# --- Models ---
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, VotingRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor, XGBClassifier
from lightgbm import LGBMRegressor, LGBMClassifier
from catboost import CatBoostRegressor, CatBoostClassifier
from sklearn.linear_model import LinearRegression, HuberRegressor
from sklearn.metrics import r2_score, f1_score, precision_score, recall_score, accuracy_score, roc_auc_score, confusion_matrix, mean_squared_error, mean_absolute_error

# Global Font Settings for Plots
plt.rcParams.update({'font.size': 14, 'axes.labelsize': 14, 'axes.titlesize': 16})

# ==============================================================================
# STEP 1: DATA LOADING & PREPROCESSING
# ==============================================================================
print("\n[STEP 1] Data Loading and Preprocessing")
print("Please upload 'KOYCEGIZ_PRECP (1).csv'...")
uploaded = files.upload()

if not uploaded:
    print("\nFile not selected. Halting execution.")
else:
    start_time = time.time()
    file_name = next(iter(uploaded))
    df_initial = pd.read_csv(io.BytesIO(uploaded[file_name]), delimiter=';')
    df_initial.columns = df_initial.columns.str.strip()
    df_initial['Date'] = df_initial.apply(lambda row: datetime(int(row['YEAR']), 1, 1) + timedelta(days=int(row['DOY']) - 1), axis=1)
    df_initial.set_index('Date', inplace=True)
    df_initial.rename(columns={'PRECIPITATION': 'Precipitation'}, inplace=True)
    print("Data loaded successfully.")

    # ==============================================================================
    # STEP 2: FEATURE ENGINEERING & DATASET CREATION
    # ==============================================================================
    print("\n[STEP 2] Creating Datasets & Calculating Statistics")

    # --- Univariate Analysis ---
    df_uni = df_initial[['Precipitation']].copy()
    for lag in [1, 3, 7, 14]:
        df_uni[f'Precipitation_lag_{lag}'] = df_uni['Precipitation'].shift(lag)
    df_uni.dropna(inplace=True)
    X_uni, y_uni = df_uni.drop('Precipitation', axis=1), df_uni['Precipitation']
    X_train_uni, X_test_uni, y_train_uni, y_test_uni = train_test_split(X_uni, y_uni, test_size=0.2, shuffle=False)
    lgbm_uni_model = LGBMRegressor(random_state=42, verbose=-1).fit(X_train_uni, y_train_uni)
    r2_uni = r2_score(y_test_uni, lgbm_uni_model.predict(X_test_uni))
    print(f"Univariate Benchmark R-squared: {r2_uni:.4f}")

    # --- Multivariate Dataset ---
    df_multi = df_initial.copy()
    TARGET = 'Precipitation'
    FEATURES = [col for col in df_multi.columns if col not in ['YEAR', 'DOY', TARGET]]
    for col in FEATURES + [TARGET]:
        for lag in [1, 3, 7]:
            df_multi[f'{col}_lag_{lag}'] = df_multi[col].shift(lag)
    df_multi.dropna(inplace=True)

    X = df_multi.drop(TARGET, axis=1)
    y = df_multi[TARGET]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # --- Data Statistics Table ---
    print("Generating Train/Test Statistics Table...")
    stats_train = y_train.describe().to_frame(name='Train Set')
    stats_test = y_test.describe().to_frame(name='Test Set')
    stats_combined = pd.concat([stats_train, stats_test], axis=1)
    display(stats_combined)
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 3: REGRESSOR OPTIMIZATION
    # ==============================================================================
    print("[STEP 3] Optimizing All 7 Regressors")

    models_to_tune = {
        "CatBoost": (CatBoostRegressor(random_state=42, verbose=0), {'iterations': [500, 1000], 'learning_rate': [0.05, 0.1], 'depth': [6, 8]}),
        "LightGBM": (LGBMRegressor(random_state=42, verbose=-1), {'n_estimators': [500, 1000], 'learning_rate': [0.05, 0.1], 'num_leaves': [31, 40]}),
        "XGBoost": (XGBRegressor(random_state=42), {'n_estimators': [500, 1000], 'learning_rate': [0.05, 0.1], 'max_depth': [6, 8]}),
        "Random Forest": (RandomForestRegressor(random_state=42), {'n_estimators': [200, 500], 'max_depth': [10, 20]}),
        "Gradient Boosting": (GradientBoostingRegressor(random_state=42), {'n_estimators': [100, 200]}),
        "Linear Regression": (LinearRegression(), {}),
        "Huber Regressor": (HuberRegressor(), {'epsilon': [1.1, 1.35], 'max_iter': [1000]})
    }

    tscv = TimeSeriesSplit(n_splits=3)
    optimized_reg_results = []
    best_estimators = {}

    for name, (model, params) in models_to_tune.items():
        if params:
            random_search = RandomizedSearchCV(estimator=model, param_distributions=params, n_iter=8, cv=tscv, scoring='r2', n_jobs=-1, random_state=42, verbose=1)
            random_search.fit(X_train_scaled, y_train)
            best_estimators[name] = random_search.best_estimator_
        else:
            model.fit(X_train_scaled, y_train)
            best_estimators[name] = model

        y_pred = best_estimators[name].predict(X_test_scaled)

        # Calculate Performance Metrics
        test_r2 = r2_score(y_test, y_pred)
        test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        test_mae = mean_absolute_error(y_test, y_pred)
        test_mse = mean_squared_error(y_test, y_pred)

        optimized_reg_results.append({
            "Model": name,
            "R2": test_r2,
            "RMSE": test_rmse,
            "MAE": test_mae,
            "MSE": test_mse
        })
        print(f"Optimization for {name} complete. R2: {test_r2:.4f}")

    optimized_reg_results_df = pd.DataFrame(optimized_reg_results).sort_values(by='R2', ascending=False)
    print("\n--- REGRESSION RESULTS ---")
    display(optimized_reg_results_df)

    # --- Ensemble Model ---
    top_3_names = optimized_reg_results_df['Model'].head(3).tolist()
    ensemble = VotingRegressor(estimators=[(name, best_estimators[name]) for name in top_3_names])
    ensemble.fit(X_train_scaled, y_train)
    y_pred_ensemble = ensemble.predict(X_test_scaled)
    r2_ensemble = r2_score(y_test, y_pred_ensemble)
    print(f"\n--- Ensemble (Top 3) Model R-squared: {r2_ensemble:.4f} ---")

    ultimate_champion_name = optimized_reg_results_df.iloc[0]['Model']
    ultimate_champion_model = best_estimators[ultimate_champion_name]
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 4: EXTREME EVENT CLASSIFICATION
    # ==============================================================================
    print("[STEP 4] Extreme Event Classification Analysis")
    extreme_threshold = df_multi[TARGET].quantile(0.95)
    print(f"   (Extreme precipitation threshold (P95): {extreme_threshold:.2f} mm/day)")
    y_cls = (df_multi[TARGET] > extreme_threshold).astype(int)
    y_train_cls, y_test_cls = y_cls[X_train.index], y_cls[X_test.index]

    scale_pos_weight = (y_train_cls == 0).sum() / (y_train_cls == 1).sum()
    classifiers = {
        "LightGBM": LGBMClassifier(random_state=42, class_weight='balanced', verbose=-1),
        "CatBoost": CatBoostClassifier(random_state=42, verbose=0, scale_pos_weight=scale_pos_weight),
        "XGBoost": XGBClassifier(random_state=42, scale_pos_weight=scale_pos_weight),
        "Random Forest": RandomForestClassifier(random_state=42, class_weight='balanced')
    }

    cls_results = []
    for name, model in classifiers.items():
        print(f"--- Evaluating Classifier: {name} ---")
        train_start_time = time.time()
        model.fit(X_train_scaled, y_train_cls)
        train_duration = time.time() - train_start_time

        y_pred_cls = model.predict(X_test_scaled)
        y_proba_cls = model.predict_proba(X_test_scaled)[:, 1]

        f1 = f1_score(y_test_cls, y_pred_cls)
        precision = precision_score(y_test_cls, y_pred_cls, zero_division=0)
        recall = recall_score(y_test_cls, y_pred_cls, zero_division=0)
        accuracy = accuracy_score(y_test_cls, y_pred_cls)
        auc = roc_auc_score(y_test_cls, y_proba_cls)

        cls_results.append({
            "Model": name, "F1-Score": f1, "AUC": auc, "Precision": precision,
            "Recall": recall, "Accuracy": accuracy, "Training Time (s)": train_duration
        })

    cls_results_df = pd.DataFrame(cls_results).sort_values(by='F1-Score', ascending=False)
    print("\n--- CLASSIFIER COMPARISON SUMMARY ---")
    display(cls_results_df.set_index('Model'))

    ultimate_classifier_name = cls_results_df.iloc[0]['Model']
    ultimate_champion_classifier = classifiers[ultimate_classifier_name]
    print(f"\nBest Performing Classifier: {ultimate_classifier_name}")
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 5: ADVANCED ANALYSIS & FORECASTING
    # ==============================================================================
    print("[STEP 5] Running Advanced Analyses")

    # --- 5a: SHAP Analysis ---
    print("--- 5a: Calculating SHAP values... ---")
    explainer = shap.TreeExplainer(ultimate_champion_model)
    shap_values = explainer.shap_values(X_test_scaled)

    # --- 5b: Extreme Value Analysis ---
    print("--- 5b: Calculating Return Periods using GEV... ---")
    annual_maxima = df_initial['Precipitation'].resample('YE').max()
    params = gev.fit(annual_maxima)
    return_periods = [2, 5, 10, 25, 50, 100]
    return_levels = [gev.ppf(1 - 1/p, *params) for p in return_periods]
    return_period_df = pd.DataFrame({'Return Period (Years)': return_periods, 'Estimated Max Daily Precipitation (mm)': return_levels})
    display(return_period_df)

    # --- 5c: Seasonal Performance Analysis ---
    print("--- 5c: Analyzing seasonal performance... ---")
    test_df_seasonal = X_test.copy()
    test_df_seasonal['Precipitation'] = y_test
    test_df_seasonal['Predicted'] = ultimate_champion_model.predict(X_test_scaled)
    test_df_seasonal['Season'] = (test_df_seasonal.index.month % 12 + 3) // 3
    seasons = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Autumn'}
    test_df_seasonal['Season'] = test_df_seasonal['Season'].map(seasons)
    seasonal_r2 = test_df_seasonal.groupby('Season').apply(lambda x: r2_score(x['Precipitation'], x['Predicted']))
    print("Seasonal R2 Scores:")
    print(seasonal_r2)

    # --- 5d: Forecasting with Prophet ---
    print("--- 5d: Forecasting to 2050 with Prophet... ---")
    prophet_df = df_initial[['Precipitation']].reset_index().rename(columns={'Date': 'ds', 'Precipitation': 'y'})
    prophet_model = Prophet().fit(prophet_df)
    future_dates = prophet_model.make_future_dataframe(periods=(2050-2025)*365)
    forecast = prophet_model.predict(future_dates)
    print("Advanced analyses complete.")
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 6: VISUALIZATION & OUTPUT GENERATION
    # ==============================================================================
    print("[STEP 6] Generating All Final Visualizations")

    files_to_download = []
    y_pred_final = ultimate_champion_model.predict(X_test_scaled)
    key_vars = ['Precipitation', 'T2M', 'T2MDEW', 'TS', 'RH2M', 'QV2M', 'WS10M']

    plt.figure(figsize=(12, 10))
    sns.heatmap(df_initial[key_vars].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap of Key Meteorological Variables')
    plt.tight_layout()
    filename = 'Correlation_Heatmap.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    plt.figure(figsize=(15, 10))
    calmap.yearplot(df_initial['Precipitation'], year=2023)
    plt.suptitle('Calendar Heatmap for Precipitation (2023 Example)', fontsize=20)
    filename = 'Calendar_Heatmap.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    plt.figure(figsize=(10, 8))
    sns.barplot(x='R2', y='Model', data=optimized_reg_results_df, palette='viridis')
    plt.axvline(x=r2_ensemble, color='red', linestyle='--', label=f'Ensemble Model R2 ({r2_ensemble:.3f})')
    plt.title('Regressor Comparison')
    plt.xlabel('R2 Score'); plt.ylabel('Model')
    plt.legend(); plt.savefig('Regressor_Comparison.png', dpi=300, bbox_inches='tight')
    files_to_download.append('Regressor_Comparison.png')
    plt.show()

    cls_results_df_for_plot = cls_results_df.set_index('Model').sort_values(by='F1-Score', ascending=True)
    fig, ax1 = plt.subplots(figsize=(12, 7))
    cls_results_df_for_plot[['F1-Score', 'AUC']].plot(kind='bar', ax=ax1, colormap='viridis', position=0, width=0.4)
    ax1.set_title('Classifier Performance: F1-Score and AUC')
    ax1.set_ylabel('Score'); ax1.set_xlabel('Model')
    ax1.tick_params(axis='x', rotation=45); ax1.grid(axis='y', linestyle='--')
    plt.legend(loc='upper left'); plt.tight_layout()
    filename = 'Classifier_Performance_Comparison.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    try:
        subset = cls_results_df[cls_results_df['Model'].isin(['CatBoost', 'LightGBM'])].copy()
        if subset.empty:
            print("Warning: CatBoost/LightGBM not found.")
        else:
            subset['Model'] = subset['Model'].map({'CatBoost': 'CatBoost C', 'LightGBM': 'LightGBM C'})
            subset = subset.sort_values('F1-Score', ascending=True)
            color_map = {'CatBoost C': '#d16d5b', 'LightGBM C': '#5b2a86'}
            colors = [color_map[m] for m in subset['Model']]

            plt.figure(figsize=(12, 8))
            plt.barh(subset['Model'], subset['F1-Score'], color=colors)
            plt.title('Classifier Comparison')
            plt.xlabel('F1-Score (Extreme Class)')
            plt.ylabel('Model')
            plt.xlim(0, 1)
            for i, v in enumerate(subset['F1-Score']):
                plt.text(v + 0.01, i, f"{v:.3f}", va='center')
            plt.tight_layout()
            filename = 'Classifier_Comparison.png'
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            files_to_download.append(filename)
            plt.show()
    except Exception as e:
        print(f"Plot error: {e}")

    plt.figure(figsize=(12, 7))
    cls_results_df.set_index('Model')['Training Time (s)'].sort_values().plot(kind='bar', color='coral')
    plt.title('Model Training Time Comparison')
    plt.ylabel('Time (seconds)'); plt.xlabel('Model')
    plt.xticks(rotation=45); plt.grid(axis='y', linestyle='--'); plt.tight_layout()
    filename = 'Classifier_Training_Time.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    y_pred_series = pd.Series(y_pred_final, index=y_test.index)
    plt.figure(figsize=(18, 9))
    plt.plot(y_train, label='Train Data (Actual)', color='dodgerblue', alpha=0.7)
    plt.plot(y_test, label='Test Data (Actual)', color='green', linewidth=2.5)
    plt.plot(y_pred_series, label=f'Prediction ({ultimate_champion_name})', color='darkorange', linestyle='--')
    plt.title('Train, Test, and Prediction Time Series')
    plt.xlabel('Date'); plt.ylabel('Precipitation (mm)')
    plt.legend(); plt.grid(True, alpha=0.5)
    plt.savefig('Train_Test_Predict.png', dpi=300)
    files_to_download.append('Train_Test_Predict.png')
    plt.show()

    plt.figure(figsize=(8, 8))
    plt.scatter(y_test, y_pred_final, alpha=0.5, c='navy')
    plt.plot([0, y_test.max()], [0, y_test.max()], color='red', linestyle='--', lw=2,
             label='Perfect Prediction (R2=1)')
    plt.title(f'Actual vs. Predicted ({ultimate_champion_name})')
    plt.xlabel('Actual Values (mm)'); plt.ylabel('Predicted Values (mm)')
    plt.legend(); plt.grid(True)
    plt.savefig('Scatter_Plot.png', dpi=300)
    files_to_download.append('Scatter_Plot.png')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.barplot(x=seasonal_r2.index, y=seasonal_r2.values, palette='coolwarm')
    plt.title('Seasonal R2 Performance')
    plt.xlabel('Season'); plt.ylabel('R-squared (R2)')
    plt.savefig('Seasonal_Performance.png', dpi=300)
    files_to_download.append('Seasonal_Performance.png')
    plt.show()

    shap.summary_plot(shap_values, X_test, plot_type="bar", show=False, max_display=25)
    plt.title('SHAP Summary Plot (Top 25 Features)')
    plt.xlabel('mean(|SHAP value|)')
    plt.tight_layout()
    plt.savefig('SHAP_Summary.png', dpi=300)
    files_to_download.append('SHAP_Summary.png')
    plt.show()

    y_pred_best_cls = ultimate_champion_classifier.predict(X_test_scaled)
    cm = confusion_matrix(y_test_cls, y_pred_best_cls)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Extreme'], yticklabels=['Normal', 'Extreme'])
    plt.title(f'Confusion Matrix ({ultimate_classifier_name})')
    plt.xlabel('Predicted Label'); plt.ylabel('True Label')
    plt.savefig('Confusion_Matrix.png', dpi=300)
    files_to_download.append('Confusion_Matrix.png')
    plt.show()

    fig_prophet = prophet_model.plot(forecast)
    ax_prophet = fig_prophet.gca()
    ax_prophet.set_title('Precipitation Forecast to 2050')
    ax_prophet.set_xlabel('Date'); ax_prophet.set_ylabel('Precipitation Trend')
    fig_prophet.tight_layout()
    filename = 'Prophet_Forecast.png'
    fig_prophet.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    # ==============================================================================
    # STEP 7: EXPORTING RESULTS & DOWNLOADING ALL ARTIFACTS
    # ==============================================================================
    print("\n[STEP 7] Exporting results to Excel and preparing for download...")

    excel_filename = 'Analysis_Results.xlsx'
    with pd.ExcelWriter(excel_filename, engine='openpyxl') as writer:
        optimized_reg_results_df.to_excel(writer, sheet_name='Regressor_Performance', index=False)
        cls_results_df.to_excel(writer, sheet_name='Classifier_Performance', index=False)
        return_period_df.to_excel(writer, sheet_name='GEV_Return_Periods', index=False)
        seasonal_r2.to_frame(name='R-squared').to_excel(writer, sheet_name='Seasonal_Performance')
        stats_combined.to_excel(writer, sheet_name='Data_Statistics')

    files_to_download.append(excel_filename)
    print(f"All result tables have been saved to '{excel_filename}' in the Colab environment.")

    print("\nDownloading all files to your computer's 'Downloads' folder...")
    for file_to_download in files_to_download:
        print(f"  Downloading: {file_to_download}")
        files.download(file_to_download)
        time.sleep(1)

    end_time = time.time()
    total_duration = end_time - start_time
    minutes = int(total_duration // 60)
    seconds = int(total_duration % 60)

    print("\n" + "="*80)
    print(f"Total execution time: {minutes} minutes and {seconds} seconds.")
    print("="*80)

    print("\nExecution complete. All artifacts have been generated, shown in Colab, and downloaded to your computer.")
    print("Process finished successfully.")
```
