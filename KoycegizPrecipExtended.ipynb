```python
# ==============================================================================
# BILINGUAL PRECIPITATION FORECASTING PIPELINE (Google Colab) - FINAL VERSION
# ÇİFT DİLLİ YAĞIŞ TAHMİNİ BORU HATTI (Google Colab) - NİHAİ VERSİYON
# ==============================================================================
# Author / Yazar: Dr. Tevfik Denizhan MUFTUOGLU
#
# License / Lisans:
# This code is released under the MIT License. You are free to use, modify,
# and distribute it, provided the original author attribution is maintained.
# ==============================================================================

# ==============================================================================
# STEP 0: INSTALLING & IMPORTING LIBRARIES
# ADIM 0: KÜTÜPHANELERİN KURULUMU VE İÇE AKTARILMASI
# ==============================================================================
print("[STEP 0] Installing and Importing All Necessary Libraries / [ADIM 0] Gerekli Tüm Kütüphaneler Kuruluyor ve İçe Aktarılıyor")
# ADDED openpyxl FOR GUARANTEED EXCEL EXPORT / GARANTİLİ EXCEL AKTARIMI İÇİN openpyxl EKLENDİ
!pip install -q prophet catboost pymannkendall shap calmap openpyxl

# --- Core Libraries / Temel Kütüphaneler ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from google.colab import files
import io
from IPython.display import display
import warnings
import time
warnings.filterwarnings("ignore")

# --- Advanced Analysis Libraries / İleri Analiz Kütüphaneleri ---
from prophet import Prophet
import shap
from scipy.stats import genextreme as gev
import calmap

# --- Scikit-learn & ML Model Libraries / Scikit-learn & ML Model Kütüphaneleri ---
from sklearn.model_selection import train_test_split, RandomizedSearchCV, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, VotingRegressor
from xgboost import XGBRegressor, XGBClassifier
from lightgbm import LGBMRegressor, LGBMClassifier
from catboost import CatBoostRegressor, CatBoostClassifier
from sklearn.metrics import r2_score, f1_score, precision_score, recall_score, accuracy_score, roc_auc_score, confusion_matrix
import pymannkendall as mk

# ==============================================================================
# STEP 1: DATA LOADING & PREPROCESSING
# ADIM 1: VERİ YÜKLEME VE ÖN İŞLEME
# ==============================================================================
print("\n[STEP 1] Data Loading and Preprocessing / [ADIM 1] Veri Yükleme ve Ön İşleme")
print("Awaiting file upload for 'KOYCEGIZ_PRECP (1).csv'... / 'KOYCEGIZ_PRECP (1).csv' dosyası için yükleme bekleniyor...")
uploaded = files.upload()

if not uploaded:
    print("\nFile not selected. Halting execution. / Dosya seçilmedi. İşlem durduruluyor.")
else:
    start_time = time.time() # <<< ZAMANLAYICI BAŞLATILDI
    # --- Data Ingestion and Preprocessing / Veri Alımı ve Ön İşleme ---
    file_name = next(iter(uploaded))
    df_initial = pd.read_csv(io.BytesIO(uploaded[file_name]), delimiter=';')
    df_initial.columns = df_initial.columns.str.strip()
    df_initial['Date'] = df_initial.apply(lambda row: datetime(int(row['YEAR']), 1, 1) + timedelta(days=int(row['DOY']) - 1), axis=1)
    df_initial.set_index('Date', inplace=True)
    df_initial.rename(columns={'PRECIPITATION': 'Precipitation'}, inplace=True)
    print("Data loaded and preprocessed. / Veri yüklendi ve ön işlendi.")

    # ==============================================================================
    # STEP 2: FEATURE ENGINEERING & DATASET CREATION
    # ADIM 2: ÖZELLİK MÜHENDİSLİĞİ VE VERİ SETİ OLUŞTURMA
    # ==============================================================================
    print("\n[STEP 2] Creating Datasets for Univariate and Multivariate Analysis / [ADIM 2] Tek Değişkenli ve Çok Değişkenli Analiz için Veri Setleri Oluşturuluyor")

    # --- Univariate Analysis (Benchmark) / Tek Değişkenli Analiz (Referans) ---
    df_uni = df_initial[['Precipitation']].copy()
    for lag in [1, 3, 7, 14]:
        df_uni[f'Precipitation_lag_{lag}'] = df_uni['Precipitation'].shift(lag)
    df_uni.dropna(inplace=True)
    X_uni, y_uni = df_uni.drop('Precipitation', axis=1), df_uni['Precipitation']
    X_train_uni, X_test_uni, y_train_uni, y_test_uni = train_test_split(X_uni, y_uni, test_size=0.2, shuffle=False)
    lgbm_uni_model = LGBMRegressor(random_state=42).fit(X_train_uni, y_train_uni)
    r2_uni = r2_score(y_test_uni, lgbm_uni_model.predict(X_test_uni))
    print(f"Univariate Benchmark R-squared (using only precipitation data): {r2_uni:.4f}")
    print(f"(TR: Sadece yağış verisi kullanan tek değişkenli referans R-kare değeri: {r2_uni:.4f})")

    # --- Multivariate Dataset Creation / Çok Değişkenli Veri Seti Oluşturma ---
    df_multi = df_initial.copy()
    TARGET = 'Precipitation'
    FEATURES = [col for col in df_multi.columns if col not in ['YEAR', 'DOY', TARGET]]
    for col in FEATURES + [TARGET]:
        for lag in [1, 3, 7]:
            df_multi[f'{col}_lag_{lag}'] = df_multi[col].shift(lag)
    df_multi.dropna(inplace=True)

    # --- Train-Test Split and Scaling / Eğitim-Test Ayrımı ve Ölçeklendirme ---
    X = df_multi.drop(TARGET, axis=1)
    y = df_multi[TARGET]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    print("Multivariate datasets created. / Çok değişkenli veri setleri oluşturuldu.")
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 3: REGRESSOR OPTIMIZATION
    # ADIM 3: REGRESYON MODELLERİNİN OPTİMİZASYONU
    # ==============================================================================
    print("[STEP 3] Optimizing Top 4 Regressors / [ADIM 3] En İyi 4 Regresyon Modelinin Optimizasyonu")
    print("Note: This step is computationally intensive and may take significant time. / Not: Bu adım yoğun hesaplama gerektirir ve önemli ölçüde zaman alabilir.")

    models_to_tune = {
        "CatBoost": (CatBoostRegressor(random_state=42, verbose=0), {'iterations': [500, 1000], 'learning_rate': [0.05, 0.1], 'depth': [6, 8]}),
        "LightGBM": (LGBMRegressor(random_state=42), {'n_estimators': [500, 1000], 'learning_rate': [0.05, 0.1], 'num_leaves': [31, 40]}),
        "XGBoost": (XGBRegressor(random_state=42), {'n_estimators': [500, 1000], 'learning_rate': [0.05, 0.1], 'max_depth': [6, 8]}),
        "Random Forest": (RandomForestRegressor(random_state=42), {'n_estimators': [200, 500], 'max_depth': [10, 20]})
    }

    tscv = TimeSeriesSplit(n_splits=3)
    optimized_reg_results = []
    best_estimators = {}

    for name, (model, params) in models_to_tune.items():
        print(f"--- Optimizing: {name} / Optimizasyon: {name} ---")
        random_search = RandomizedSearchCV(estimator=model, param_distributions=params, n_iter=8, cv=tscv, scoring='r2', n_jobs=-1, random_state=42, verbose=1)
        random_search.fit(X_train_scaled, y_train)

        best_estimators[name] = random_search.best_estimator_
        y_pred = best_estimators[name].predict(X_test_scaled)
        test_r2 = r2_score(y_test, y_pred)
        optimized_reg_results.append({"Model": name, "Optimized R²": test_r2})
        print(f"Optimization for {name} complete. Best CV R²: {random_search.best_score_:.4f}. Test Set R²: {test_r2:.4f}")

    optimized_reg_results_df = pd.DataFrame(optimized_reg_results).sort_values(by='Optimized R²', ascending=False)
    print("\n--- FINAL RESULTS (OPTIMIZED REGRESSORS) ---")
    print("--- (TR) FİNAL SONUÇLARI (OPTİMİZE EDİLMİŞ MODELLER) ---")
    display(optimized_reg_results_df)

    # --- Ensemble Model (Voting Regressor) / Topluluk Modeli ---
    top_3_names = optimized_reg_results_df['Model'].head(3).tolist()
    ensemble = VotingRegressor(estimators=[(name, best_estimators[name]) for name in top_3_names])
    ensemble.fit(X_train_scaled, y_train)
    y_pred_ensemble = ensemble.predict(X_test_scaled)
    r2_ensemble = r2_score(y_test, y_pred_ensemble)
    print(f"\n--- Ensemble (Top 3 Optimized) Model R-squared: {r2_ensemble:.4f} ---")
    print(f"--- (TR) Topluluk (En İyi 3) Model R-kare Değeri: {r2_ensemble:.4f} ---")

    ultimate_champion_name = optimized_reg_results_df.iloc[0]['Model']
    ultimate_champion_model = best_estimators[ultimate_champion_name]
    print(f"\nBest Performing Regressor / En İyi Performans Gösteren Regresyon Modeli: {ultimate_champion_name}")
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 4: EXTREME EVENT CLASSIFICATION (COMPREHENSIVE VERSION)
    # ADIM 4: AŞIRI OLAY SINIFLANDIRMASI (KAPSAMLI VERSİYON)
    # ==============================================================================
    print("[STEP 4] Extreme Event Classification Analysis / [ADIM 4] Aşırı Olay Sınıflandırma Analizi")
    extreme_threshold = df_multi[TARGET].quantile(0.95)
    print(f"   (Extreme precipitation threshold (P95): {extreme_threshold:.2f} mm/day / Aşırı yağış eşiği (P95): {extreme_threshold:.2f} mm/gün)")
    y_cls = (df_multi[TARGET] > extreme_threshold).astype(int)
    y_train_cls, y_test_cls = y_cls[X_train.index], y_cls[X_test.index]

    scale_pos_weight = (y_train_cls == 0).sum() / (y_train_cls == 1).sum()
    classifiers = {
        "LightGBM": LGBMClassifier(random_state=42, class_weight='balanced'),
        "CatBoost": CatBoostClassifier(random_state=42, verbose=0, scale_pos_weight=scale_pos_weight),
        "XGBoost": XGBClassifier(random_state=42, scale_pos_weight=scale_pos_weight),
        "Random Forest": RandomForestClassifier(random_state=42, class_weight='balanced')
    }

    cls_results = []
    for name, model in classifiers.items():
        print(f"--- Evaluating Classifier: {name} / Sınıflandırıcı Değerlendiriliyor: {name} ---")
        train_start_time = time.time()
        model.fit(X_train_scaled, y_train_cls)
        train_duration = time.time() - train_start_time

        y_pred_cls = model.predict(X_test_scaled)
        y_proba_cls = model.predict_proba(X_test_scaled)[:, 1]

        f1 = f1_score(y_test_cls, y_pred_cls)
        precision = precision_score(y_test_cls, y_pred_cls, zero_division=0)
        recall = recall_score(y_test_cls, y_pred_cls, zero_division=0)
        accuracy = accuracy_score(y_test_cls, y_pred_cls)
        auc = roc_auc_score(y_test_cls, y_proba_cls)

        cls_results.append({
            "Model": name, "F1-Score": f1, "AUC": auc, "Precision": precision,
            "Recall": recall, "Accuracy": accuracy, "Training Time (s)": train_duration
        })

    cls_results_df = pd.DataFrame(cls_results).sort_values(by='F1-Score', ascending=False)
    print("\n--- COMPREHENSIVE CLASSIFIER COMPARISON SUMMARY ---")
    print("--- (TR) KAPSAMLI SINIFLANDIRICI KARŞILAŞTIRMA ÖZETİ ---")
    display(cls_results_df.set_index('Model'))

    ultimate_classifier_name = cls_results_df.iloc[0]['Model']
    ultimate_champion_classifier = classifiers[ultimate_classifier_name]
    print(f"\nBest Performing Classifier / En İyi Performans Gösteren Sınıflandırıcı: {ultimate_classifier_name}")
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 5: ADVANCED ANALYSIS & FORECASTING
    # ADIM 5: İLERİ ANALİZ VE TAHMİN
    # ==============================================================================
    print("[STEP 5] Running Advanced Analyses / [ADIM 5] İleri Analizler Yürütülüyor")

    # --- 5a: SHAP Analysis ---
    print("--- 5a: Calculating SHAP values... / SHAP değerleri hesaplanıyor... ---")
    explainer = shap.TreeExplainer(ultimate_champion_model)
    shap_values = explainer.shap_values(X_test_scaled)

    # --- 5b: Extreme Value Analysis ---
    print("--- 5b: Calculating Return Periods using GEV... / GEV ile Tekerrür Periyotları hesaplanıyor... ---")
    annual_maxima = df_initial['Precipitation'].resample('YE').max()
    params = gev.fit(annual_maxima)
    return_periods = [2, 5, 10, 25, 50, 100]
    return_levels = [gev.ppf(1 - 1/p, *params) for p in return_periods]
    return_period_df = pd.DataFrame({'Return Period (Years)': return_periods, 'Estimated Max Daily Precipitation (mm)': return_levels})
    print("Return Period Analysis Results / Tekerrür Periyodu Analiz Sonuçları:")
    display(return_period_df)

    # --- 5c: Seasonal Performance Analysis ---
    print("--- 5c: Analyzing seasonal performance... / Mevsimsel performans analiz ediliyor... ---")
    test_df_seasonal = X_test.copy()
    test_df_seasonal['Precipitation'] = y_test
    test_df_seasonal['Predicted'] = ultimate_champion_model.predict(X_test_scaled)
    test_df_seasonal['Season'] = (test_df_seasonal.index.month % 12 + 3) // 3
    seasons = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Autumn'}
    test_df_seasonal['Season'] = test_df_seasonal['Season'].map(seasons)
    seasonal_r2 = test_df_seasonal.groupby('Season').apply(lambda x: r2_score(x['Precipitation'], x['Predicted']))
    print("Seasonal R² Scores / Mevsimsel R² Skorları:")
    print(seasonal_r2)

    # --- 5d: Forecasting with Prophet ---
    print("--- 5d: Forecasting to 2050 with Prophet... / Prophet ile 2050'ye tahmin yapılıyor... ---")
    prophet_df = df_initial[['Precipitation']].reset_index().rename(columns={'Date': 'ds', 'Precipitation': 'y'})
    prophet_model = Prophet().fit(prophet_df)
    future_dates = prophet_model.make_future_dataframe(periods=(2050-2025)*365)
    forecast = prophet_model.predict(future_dates)
    print("Advanced analyses complete. / İleri analizler tamamlandı.")
    print("\n" + "="*80 + "\n")

    # ==============================================================================
    # STEP 6: VISUALIZATION & OUTPUT GENERATION
    # ADIM 6: GÖRSELLEŞTİRME VE ÇIKTI OLUŞTURMA
    # ==============================================================================
    print("[STEP 6] Generating All Final Visualizations / [ADIM 6] Tüm Nihai Görseller Oluşturuluyor")

    files_to_download = []
    y_pred_final = ultimate_champion_model.predict(X_test_scaled)
    key_vars = ['Precipitation', 'T2M', 'T2MDEW', 'TS', 'RH2M', 'QV2M', 'WS10M']

    # --- Visualizations (All plots will be shown in Colab and saved for download) ---
    plt.figure(figsize=(12, 10))
    sns.heatmap(df_initial[key_vars].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap of Key Meteorological Variables\n(Temel Meteorolojik Değişkenlerin Korelasyon Isı Haritası)', fontsize=16)
    plt.tight_layout()
    filename = 'Correlation_Heatmap.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    plt.figure(figsize=(15, 10))
    calmap.yearplot(df_initial['Precipitation'], year=2023)
    plt.suptitle('Calendar Heatmap for Precipitation (2023 Example)\n(Yağış için Takvim Isı Haritası - 2023 Örneği)', fontsize=20)
    filename = 'Calendar_Heatmap.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    plt.figure(figsize=(10, 8))
    sns.barplot(x='Optimized R²', y='Model', data=optimized_reg_results_df, palette='viridis')
    plt.axvline(x=r2_ensemble, color='red', linestyle='--', label=f'Ensemble Model R² ({r2_ensemble:.3f})')
    plt.title('Optimized Regressor Comparison\n(Optimize Edilmiş Regresyon Modeli Karşılaştırması)', fontsize=16)
    plt.xlabel('Optimized R² Score / Optimize R² Skoru'); plt.ylabel('Model / Model')
    plt.legend(); plt.savefig('Regressor_Comparison.png', dpi=300, bbox_inches='tight')
    files_to_download.append('Regressor_Comparison.png')
    plt.show()

    # (A) F1 & AUC for all classifiers (kept)
    cls_results_df_for_plot = cls_results_df.set_index('Model').sort_values(by='F1-Score', ascending=True)
    fig, ax1 = plt.subplots(figsize=(12, 7))
    cls_results_df_for_plot[['F1-Score', 'AUC']].plot(kind='bar', ax=ax1, colormap='viridis', position=0, width=0.4)
    ax1.set_title('Classifier Performance: F1-Score and AUC\n(Sınıflandırıcı Performansı: F1-Skoru ve AUC)', fontsize=16)
    ax1.set_ylabel('Score / Skor', fontsize=12); ax1.set_xlabel('Model / Model', fontsize=12)
    ax1.tick_params(axis='x', rotation=45); ax1.grid(axis='y', linestyle='--')
    plt.legend(loc='upper left'); plt.tight_layout()
    filename = 'Classifier_Performance_Comparison.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    # (B) EXACT "Classifier Comparison" — bilingual + distinct colors (horizontal F1 for CatBoost C & LightGBM C)
    try:
        subset = cls_results_df[cls_results_df['Model'].isin(['CatBoost', 'LightGBM'])].copy()
        if subset.empty:
            print("Uyarı/Warning: CatBoost ve LightGBM sonuçları bulunamadı; 'Classifier_Comparison' atlandı.")
        else:
            subset['Model'] = subset['Model'].map({'CatBoost': 'CatBoost C', 'LightGBM': 'LightGBM C'})
            subset = subset.sort_values('F1-Score', ascending=True)
            color_map = {'CatBoost C': '#d16d5b', 'LightGBM C': '#5b2a86'}  # farklı ve güçlü renkler / vivid distinct colors
            colors = [color_map[m] for m in subset['Model']]

            plt.figure(figsize=(12, 8))
            plt.barh(subset['Model'], subset['F1-Score'], color=colors)
            plt.title('Classifier Comparison\n(Sınıflandırıcı Karşılaştırması)')
            plt.xlabel('F1-Score (Extreme Class) / F1-Skoru (Aşırı Sınıf)')
            plt.ylabel('Model / Model')
            plt.xlim(0, 1)
            for i, v in enumerate(subset['F1-Score']):
                plt.text(v + 0.01, i, f"{v:.3f}", va='center')  # değer etiketi / value label
            plt.tight_layout()
            filename = 'Classifier_Comparison.png'
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            files_to_download.append(filename)
            plt.show()
    except Exception as e:
        print(f"'Classifier Comparison' plot could not be created / Grafik oluşturulamadı: {e}")

    plt.figure(figsize=(12, 7))
    cls_results_df.set_index('Model')['Training Time (s)'].sort_values().plot(kind='bar', color='coral')
    plt.title('Model Training Time Comparison\n(Model Eğitim Süresi Karşılaştırması)', fontsize=16)
    plt.ylabel('Time (seconds) / Süre (saniye)', fontsize=12); plt.xlabel('Model / Model', fontsize=12)
    plt.xticks(rotation=45); plt.grid(axis='y', linestyle='--'); plt.tight_layout()
    filename = 'Classifier_Training_Time.png'
    plt.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    y_pred_series = pd.Series(y_pred_final, index=y_test.index)
    plt.figure(figsize=(18, 9))
    plt.plot(y_train, label='Train Data (Actual) / Eğitim Verisi (Gerçek)', color='dodgerblue', alpha=0.7)
    plt.plot(y_test, label='Test Data (Actual) / Test Verisi (Gerçek)', color='green', linewidth=2.5)
    plt.plot(y_pred_series, label=f'Prediction ({ultimate_champion_name}) / Tahmin', color='darkorange', linestyle='--')
    plt.title('Train, Test, and Prediction Time Series\n(Eğitim, Test ve Tahmin Zaman Serisi)', fontsize=18)
    plt.xlabel('Date / Tarih'); plt.ylabel('Precipitation (mm) / Yağış (mm)')
    plt.legend(); plt.grid(True, alpha=0.5)
    plt.savefig('Train_Test_Predict.png', dpi=300)
    files_to_download.append('Train_Test_Predict.png')
    plt.show()

    plt.figure(figsize=(8, 8))
    plt.scatter(y_test, y_pred_final, alpha=0.5, c='navy')
    plt.plot([0, y_test.max()], [0, y_test.max()], color='red', linestyle='--', lw=2,
             label='Perfect Prediction (R²=1)\nMükemmel Tahmin (R²=1)')
    plt.title(f'Actual vs. Predicted ({ultimate_champion_name})\n(Gerçek vs. Tahmin)', fontsize=16)
    plt.xlabel('Actual Values (mm) / Gerçek Değerler (mm)')
    plt.ylabel('Predicted Values (mm) / Tahmin Edilen Değerler (mm)')
    plt.legend(); plt.grid(True)
    plt.savefig('Scatter_Plot.png', dpi=300)
    files_to_download.append('Scatter_Plot.png')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.barplot(x=seasonal_r2.index, y=seasonal_r2.values, palette='coolwarm')
    plt.title('Seasonal R² Performance\n(Mevsimsel R² Performansı)', fontsize=16)
    plt.xlabel('Season / Mevsim'); plt.ylabel('R-squared (R²)')
    plt.savefig('Seasonal_Performance.png', dpi=300)
    files_to_download.append('Seasonal_Performance.png')
    plt.show()

    shap.summary_plot(shap_values, X_test, plot_type="bar", show=False, max_display=25)
    plt.title('SHAP Summary Plot (Top 25 Features)\n(SHAP Özet Grafiği - En İyi 25 Özellik)', fontsize=16)
    plt.xlabel('mean(|SHAP value|) (average impact on model output magnitude)\n(ortalama etki)')
    plt.tight_layout()
    plt.savefig('SHAP_Summary.png', dpi=300)
    files_to_download.append('SHAP_Summary.png')
    plt.show()

    y_pred_best_cls = ultimate_champion_classifier.predict(X_test_scaled)
    cm = confusion_matrix(y_test_cls, y_pred_best_cls)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Extreme'], yticklabels=['Normal', 'Extreme'])
    plt.title(f'Confusion Matrix ({ultimate_classifier_name})\n(Karışıklık Matrisi)', fontsize=16)
    plt.xlabel('Predicted Label / Tahmin Edilen Etiket'); plt.ylabel('True Label / Gerçek Etiket')
    plt.savefig('Confusion_Matrix.png', dpi=300)
    files_to_download.append('Confusion_Matrix.png')
    plt.show()

    fig_prophet = prophet_model.plot(forecast)
    ax_prophet = fig_prophet.gca()
    ax_prophet.set_title('Precipitation Forecast to 2050\n(2050 Yılına Kadar Yağış Tahmini)', fontsize=18)
    ax_prophet.set_xlabel('Date / Tarih'); ax_prophet.set_ylabel('Precipitation Trend / Yağış Trendi')
    fig_prophet.tight_layout()
    filename = 'Prophet_Forecast.png'
    fig_prophet.savefig(filename, dpi=300)
    files_to_download.append(filename)
    plt.show()

    # ==============================================================================
    # STEP 7: EXPORTING RESULTS & DOWNLOADING ALL ARTIFACTS
    # ADIM 7: SONUÇLARIN DIŞARI AKTARILMASI VE TÜM ÇIKTILARIN İNDİRİLMESİ
    # ==============================================================================
    print("\n[STEP 7] Exporting results to Excel and preparing for download... / [ADIM 7] Sonuçlar Excel'e aktarılıyor ve indirmeye hazırlanıyor...")

    # --- Create an Excel file with all result tables in Colab's environment ---
    excel_filename = 'Analysis_Results.xlsx'
    with pd.ExcelWriter(excel_filename, engine='openpyxl') as writer:
        optimized_reg_results_df.to_excel(writer, sheet_name='Regressor_Performance', index=False)
        cls_results_df.to_excel(writer, sheet_name='Classifier_Performance', index=False)
        return_period_df.to_excel(writer, sheet_name='GEV_Return_Periods', index=False)
        seasonal_r2.to_frame(name='R-squared').to_excel(writer, sheet_name='Seasonal_Performance')

    files_to_download.append(excel_filename)
    print(f"All result tables have been saved to '{excel_filename}' in the Colab environment.")
    print(f"Tüm sonuç tabloları Colab ortamındaki '{excel_filename}' dosyasına kaydedildi.")

    # --- Download all artifacts (images + Excel file) to your local computer's "Downloads" folder ---
    print("\nDownloading all files to your computer's 'Downloads' folder...")
    print("Tüm dosyalar bilgisayarınızın 'İndirilenler' klasörüne indiriliyor...")
    for file_to_download in files_to_download:
        print(f"  Downloading / İndiriliyor: {file_to_download}")
        files.download(file_to_download)
        time.sleep(1)  # Tarayıcı engellemesini önlemek için indirme istekleri arasına 1 saniye gecikme eklendi

    # <<< TOPLAM ÇALIŞMA SÜRESİNİ HESAPLA VE YAZDIR
    end_time = time.time()
    total_duration = end_time - start_time
    minutes = int(total_duration // 60)
    seconds = int(total_duration % 60)

    print("\n" + "="*80)
    print(f"Total execution time: {minutes} minutes and {seconds} seconds.")
    print(f"Toplam çalışma süresi: {minutes} dakika {seconds} saniye.")
    print("="*80)

    print("\nExecution complete. All artifacts have been generated, shown in Colab, and downloaded to your computer.")
    print("İşlem tamamlandı. Tüm çıktılar oluşturuldu, Colab'de gösterildi ve bilgisayarınıza indirildi.")
```
